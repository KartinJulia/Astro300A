Textbook problem 7.7

This problem is investigating the vertical motion of the Sun in the galaxy.  The first thing we need to do is to find the constants $A$ and $B$ that solve the equation

$$z(t)=A\sin(\nu t)+B\cos(\nu(t))$$

We know that the Sun is currently at 25 pc above the midplane and is moving at 7km s$^{-1}$ upward.  We also know that $\nu=\sqrt{4\pi G \rho}$ where $\rho=0.1$M$_\odot$ pc$^{-3}$.

We can solve for $A$ and $B$ by setting $t=0$ for the present day.  Thus,

$$ 
\begin{aligned}
z(t=0)=&A\sin(\nu t)+B\cos(\nu(t))\\
25\textrm{pc} =& A\sin(0)+B\cos(0))\\
25\textrm{pc} =& B
\end{aligned}
$$

Similarly, we can find the derivative of $z(t)$ to find $A$.

$$ 
\begin{aligned}
\dot{z}(t=0)=&A\nu\cos(\nu t)-B\nu\sin(\nu(t))\\
7\textrm{km s}^{-1} =& A\nu\cos(0)-B\nu\sin(0))\\
7\textrm{km s}^{-1} =& A\nu
\end{aligned}
$$

These two equations give us $A=95.2$ pc and $B=25$ pc.  I'll use numerical methods to find the last midplane crossing in the program below, but to spoil the answer, $t_{z=0}=-3.42$ Myr.

Below is my code to plot the position as a function of time and to find the last crossing.

```python
# Import packages
from pylab import *
import math,numpy,matplotlib
%matplotlib inline
```

```python
# Set constants

# cm in a pc
pc = 3.086E18 # cm

# s to Myr
myr = pi*1E7*1E6  # s--pi*10^7 s/yr, 10^6 yr/Myr

# Gravitational constant
G = 6.674E-8 # cm^3/g/s^2

# Density
rho = 6.77E-24 #g/cm^3

# Frequency
nu = sqrt(4.*pi*G*rho) #1/s

# Distance and velocity at t=0 
d = 25.*pc #cm
v = 7.E5 #cm/s
```

```python
# Now create lambda functions z(A,B,t) and dzdt(A,B,t) to make life easier
z = lambda A,B,t: A*sin(nu*t)+B*cos(nu*t)
dzdt = lambda A,B,t: A*nu*cos(nu*t)-B*nu*sin(nu*t)
```

```python
# Create an array of time values
# remember linspace takes min, max, and number of values in array
t = linspace(-5E15,5E15,1E5) #

# Calculate constants A and B
A = v/nu
B = d
print("A =",A/pc,"; B =",B/pc) # print out A and B in pc
```

```python
# Make a plot!  I'm plotting both position and velocity, so I'm using two subplots

# set up my figure and subplots
f=figure(figsize=(7,10))
from matplotlib import gridspec
gs = gridspec.GridSpec(2, 1, height_ratios=[1, 1]) 
ax=subplot(gs[0])

# Plotting z(t) where t is in Myr and z is in pc. The t array I created is in seconds, 
# so I convert to Myr.
plot(t/myr,z(A,B,t)/pc)
# Plot the current day value
plot(t/myr,ones_like(t)*d/pc,'r',lw=1)
# Plot t=0
plot([0,0],[-110,110],'k:')

# label my y axis
ylabel('Vertical position (pc)') 

# Set up nice plotting limits (+/- 110 pc in y and +/- 150 Myr in x)
ylim(-110,110)
xlim(-1.5E2,1.5E2)

## Find when the last crossing occurred 

# Let's create a smaller search range.  We know it should have occurred within the last 10 Myr/
tsearch=linspace(-10*myr,0,5E5)
# We can use a where statment to find the index where z(tsearch) was at a minimum
crossing=where(abs(z(A,B,tsearch)) == min(abs(z(A,B,tsearch))))[0]
# then, we take the value of t where z was a minimum
t0=tsearch[crossing]
print("Time of last crossing in Myr:",t0/myr)
# Plot this point with a magenta star
plot(t0/myr,0,'m*',ms=20)

# set the x-axis to invisible so it doesn't interfere with my velocity plot
plt.setp(ax.get_xticklabels(), visible=False)

# make the velocity plot
ax1=subplot(gs[1],sharex=ax)
plot(t/myr,dzdt(A,B,t)/1E5)
plot(t/myr,ones_like(t)*v/1E5,'r',lw=1)
plot([0,0],[-10,10],'k:')

# set nice limits
ylim(-9,9)
xlim(-1.5E2,1.5E2)

# label x and y axes
ylabel('Vertical velocity (km/s)') 
xlabel('Time (Myr)')

# make my subplots share the x axis
plt.subplots_adjust(hspace=.0)
```
